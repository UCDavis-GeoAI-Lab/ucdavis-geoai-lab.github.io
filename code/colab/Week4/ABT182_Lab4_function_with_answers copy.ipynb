In this exercise, we will write a Python function that converts geographic coordinates from  
Degrees–Minutes–Seconds (DMS) format to Decimal Degrees (DD) format.

### Objective
Create a function named `dms_to_dd()` that converts latitude or longitude values from DMS to decimal degrees.

### Conversion Formula
$$
\text{Decimal Degrees} = \text{Degrees} + \frac{\text{Minutes}}{60} + \frac{\text{Seconds}}{3600}
$$

### Instructions
Your function should:

1. Take three input arguments:
   - degrees
   - minutes
   - seconds
2. Convert the input values from DMS format to decimal degrees.
3. Return the decimal degree value.
4. Include a docstring using triple quotes (`""" """`) that explains:
   - What the function does
   - The input parameters
   - The returned value
5. Call the function to convert **36 degrees, 7 minutes, and 30 seconds**.
6. Print the result to the output.

Resources:
- [Converting Degrees, Minutes, and Seconds to Decimal Degrees](https://study.com/skill/learn/how-to-convert-degrees-minutes-seconds-to-decimal-degrees-explanation.html)
- [Compare your results with this convertor](https://www.fcc.gov/media/radio/dms-decimal)


````python
# write your code here

def dms_to_dd(degrees, minutes, seconds):
    """Converts DMS (Degrees, Minutes, Seconds) to Decimal Degrees."""
    return degrees + (minutes/60) + (seconds/3600)

# Example usage
print(dms_to_dd(36, 7, 30))  # Example for 36 degrees, 7 minutes, 30 seconds
````


````python
# write your code here

def filter_point_data(data, bbox):
    """
    Filters GIS data to include only points within a given bounding box.

    input:
        - data is a dictionary
        - bbox is a tuple with four points defining southwest (lower left) and northeast (upper right) corners.

    output:
        - a dictionary containing the points inside the bbox.

    """

    filtered_data = {}

    for p in data:
        point = data[p]

        lat_in_range = bbox[0] <= point[0] <= bbox[2]
        lon_in_range = bbox[1] <= point[1] <= bbox[3]

        if lat_in_range and lon_in_range:
            filtered_data[p] = point

    return filtered_data


````

````python
# Part A
# Call your function
point_dic = {
            'p1': (40.7189, -73.9981),  # Inside the Bounding Box
            'p2': (40.7292, -73.9904),  # Inside the Bounding Box
            'p3': (40.6890, -73.9977),  # Outside the Bounding Box
            'p4': (40.7065, -73.9795),  # Outside the Bounding Box
            'p5': (40.6580, -74.0532),  # Outside the Bounding Box
            "p6": (40.7128, -74.0060)
            }

bbox = (40.7000, -74.0200, 40.7300, -73.9900)

pnt_in_bbox = filter_point_data(point_dic, bbox)
print(pnt_in_bbox)
````

- Given a NumPy array representing yearly average temperatures for the last 24 years from 2000 to 2024, write a function that returns the years (e.g., 2021) with temperature anomalies. In this case, an anomaly is defined as a temperature that is more than **two** standard deviations away from the mean.

- The function should return:
  - A dictionary where:
    - keys are the years (e.g., `2020`)
    - values are the corresponding temperatures

- Hint: once you find the anomaly years and their temperature, you can use `zip()`,
```output = dict(zip(anomaly_years, anomaly_values))```

```
# years from 2000 to 2024
years = np.arange(2000, 2025)

# temp data for 24 years
temperatures = np.array([28.82, 22.00, 24.89, 31.20, 29.34, 10.11, 24.75, 19.24, 19.48, 22.05, 20.72, 27.27, 23.81, 15.00, 22.22, 21.67, 27.47, 18.97, 35.00, 15.73, 38.00, 23.27, 24.32, 16.29])  # Sample data
```

- Expected outcome
```
{2005: 10.11, 2020: 38.0}
```

````python
# write your code here
# feel free to add more code cells if needed.

import numpy as np

def find_anomaly_years(temperatures, years):

    """Returns the years with temperature anomalies and the corresponding temp."""

    mean = np.mean(temperatures)
    std_dev = np.std(temperatures)
    anomaly_indices = np.where(abs(temperatures - mean) > 2* std_dev)[0]

    # Years
    anomaly_years = [years[index] for index in anomaly_indices]
    anomaly_values = [temperatures[index] for index in anomaly_indices]
    output = dict(zip(anomaly_years, anomaly_values))

    return output

# Using the previously created temperatures array
years = np.arange(2000, 2025)

temperatures = np.array([28.82, 22.00, 24.89, 31.20, 29.34, 10.11, 24.75, 19.24, 19.48, 22.05, 20.72, 27.27, 23.81, 15.00, 22.22, 21.67, 27.47, 18.97, 35.00, 15.73, 38.00, 23.27, 24.32, 16.29])  # Sample data

anomaly_years = find_anomaly_years(temperatures, years)
anomaly_years
````

Write a function called `plot_custom_bar()` that:

Run your function for the number of EV charing stations for 10 cities in CA.
```
cities = ['Los Angeles', 'San Francisco', 'San Diego', 'Sacramento', 'Fresno', 'San Jose', 'Oakland', 'Long Beach', 'Bakersfield', 'Anaheim']
```

**NOTE**

- the input data (`cities` and `n_station`) are `list`, you should first find a way to make a dictionary. Hint: use `zip()`.
- City names on x-axis may appear cluttered or overlap on each other. To prevent this and make it readiable, use `plt.xticks(rotation=XX).`, and assign a good number of `XX`.


````python
# write your code here
def plot_custom_bar(data):
    colors = ['green' if value > 50 else 'red' for value in data.values()]
    plt.bar(data.keys(), data.values(), color=colors)
    plt.xlabel('Category')
    plt.ylabel('Value')
    plt.title('Customized Bar Plot')
    plt.xticks(rotation=70)
    plt.show()



````

````python
# call the function
cities = ['Los Angeles', 'San Francisco', 'San Diego', 'Sacramento', 'Fresno', 'San Jose', 'Oakland', 'Long Beach', 'Bakersfield', 'Anaheim']
n_station = [70, 85, 50, 32, 10, 62, 31, 14, 9, 17]
data = dict(zip(cities, n_station))
plot_custom_bar(data)
````

````python
# write your code here

import matplotlib.pyplot as plt

def vegetation_map(data, cmap='viridis'):
    # Separate data into x, y, types, and sizes
    x = [item[0] for item in data]
    y = [item[1] for item in data]
    types = [item[2] for item in data]
    sizes = [item[3] for item in data]  # Scale size for better visibility

    # Create a scatter plot
    plt.scatter(x, y, c=types, s=sizes, alpha=0.6, cmap=cmap )
    plt.colorbar(label='Vegetation Type')
    plt.xlabel('Longitude')
    plt.ylabel('Latitude')
    plt.title('Vegetation Types and Density in Nature Reserve')
    plt.show()
````
