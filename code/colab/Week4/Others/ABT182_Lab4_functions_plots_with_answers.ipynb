- Given a NumPy array representing yearly average temperatures for the last 24 years from 2000 to 2024, write a function that returns the years (e.g., 2021) with temperature anomalies. In this case, an anomaly is defined as a temperature that is more than **two** standard deviations away from the mean.
- The function should return the years with temperature anomalies and the corresponding temperature as a dictionary. Print the output of the function in the terminal. 

- Hint: once you find the anomaly years and their temperature, you can use `zip()`,
```output = dict(zip(anomaly_years, anomaly_values))```, and return the zipped output (which will be a dictionary). 



```
# years from 2000 to 2024
years = np.arange(2000, 2025)

# temp data for 24 years
temperatures = np.array([28.82, 22.00, 24.89, 31.20, 29.34, 10.11, 24.75, 19.24, 19.48, 22.05, 20.72, 27.27, 23.81, 15.00, 22.22, 21.67, 27.47, 18.97, 35.00, 15.73, 38.00, 23.27, 24.32, 16.29])  # Sample data
```

- Expected outcome
```
{2005: 10.11, 2020: 38.0}
```

````python
import numpy as np

def find_temperature_anomalies(years, temperatures):
    '''
    Find years with temperature anomalies (more than 2 standard deviations from mean).
    
    Parameters:
        years: Array of years
        temperatures: Array of temperatures
    
    Returns:
        Dictionary with anomaly years as keys and temperatures as values
    '''
    mean_temp = np.mean(temperatures)
    std_temp = np.std(temperatures)
    
    # Find anomalies (more than 2 std away from mean)
    threshold = 2 * std_temp
    anomaly_indices = np.where(np.abs(temperatures - mean_temp) > threshold)[0]
    
    anomaly_years = years[anomaly_indices]
    anomaly_values = temperatures[anomaly_indices]
    
    # Create dictionary using zip
    output = dict(zip(anomaly_years, anomaly_values))
    
    return output

````

````python
# Using the previously created temperatures array
years = np.arange(2000, 2025)

temperatures = np.array([28.82, 22.00, 24.89, 31.20, 29.34, 10.11, 24.75, 19.24, 19.48, 22.05, 20.72, 27.27, 23.81, 15.00, 22.22, 21.67, 27.47, 18.97, 35.00, 15.73, 38.00, 23.27, 24.32, 16.29])  # Sample data

# Call the function and print the output
anomalies = find_temperature_anomalies(years, temperatures)
print(anomalies)

````

- the name of the cities may overlap (fall on the top of each other) on the x-axis. To prevent this and make it more readable, use `plt.xticks(rotation=XX).`, assign a good number to `XX`, which will be a degree to rotate the text with respect to the x axis.

````python
import matplotlib.pyplot as plt
def plot_custom_bar(data_dict):
    '''
    Plot a bar graph with colors based on values.
    Parameters:
        data_dict: Dictionary with categories as keys and numbers as values
    '''
    categories = list(data_dict.keys())
    values = list(data_dict.values())
    # Assign colors: red for values > 50, green otherwise
    colors = ['red' if val > 50 else 'green' for val in values]
    plt.figure(figsize=(10, 6))
    plt.bar(categories, values, color=colors)
    plt.title('Custom Bar Plot')
    plt.xlabel('Categories')
    plt.ylabel('Values')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
````

````python
import matplotlib.pyplot as plt

def plot_gis_scatter(data, cmap='viridis'):
    '''
    Create a scatter plot representing vegetation locations.
    
    Parameters:
        data: List of tuples (x, y, vegetation_type, density)
        cmap: Colormap name (default: 'viridis')
    '''
    x_coords = [point[0] for point in data]
    y_coords = [point[1] for point in data]
    vegetation_types = [point[2] for point in data]
    densities = [point[3] for point in data]
    
    plt.figure(figsize=(10, 8))
    scatter = plt.scatter(x_coords, y_coords, c=densities, s=densities, 
                         cmap=cmap, alpha=0.6, edgecolors='black')
    plt.colorbar(scatter, label='Density')
    plt.xlabel('Longitude')
    plt.ylabel('Latitude')
    plt.title('GIS Scatter Plot - Vegetation Density')
    plt.grid(True, alpha=0.3)
    plt.show()


````
