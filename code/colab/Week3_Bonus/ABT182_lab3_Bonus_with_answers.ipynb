3. Check the variable type for both `math_score` and `biology_score` variables. Hint: Use Python built-in `type()` function.

4. Given the variables, print a sentence that contains both the `math_score` and `biology_score` with a **space in between**. For example:

The average score is 93.80

To practice different printing styles, use the print() function to display the same message in **three ways: 1) using commas, 2) using the + sign, and 3) using an f-string.** Compare the outputs and note their differences.

```python
# Write your code here

math_score = input("Enter your Math exam score: ")
biology_score = input("Enter your Biology exam score: ")

print(type(math_score))
print(type(biology_score))

print("My Math score is", math_score, "and my Biology score is", biology_score)

import statistics

average_score = statistics.mean([float(math_score), float(biology_score)])

print("The average score is", round(average_score, 2))
print("The average score is " + str(round(average_score, 2)))
print(f"The average score is {average_score:.2f}")
```

### **Excercise 3**: Working with Tuple and Set Variables

In this exercise, you will review **tuple** and **set** variables using names of people.  
You will observe how tuples and sets store data differently by accessing elements, checking variable types, and sorting values.

---

**Instructions**

1. Define a **tuple** called `name_tuple` with the following values:  
   `"Alex"`, `"Jane"`, `"Sam"`, and `"Sophia"`.

2. Print the **first name** in `name_tuple` in **uppercase letters**.  
   *Hint: Use `upper()` method.*

3. Define a **set** called `name_set` with the following values:  
   `"Alex"`, `"Maria"`, `"John"`, `"Maria"`, `"Sophia"`, `"Alex"`. Print `name_set` and observe what happens to the duplicate names.

4. Use the `sorted()` **function** to create a sorted version of `name_set`.  
   Store the result in a variable called `sorted_names` and print it.
   
5. Check the type of `name_tuple`, `name_set` and `sorted_names` variables.

6. Find the **common names** between `name_tuple` and `name_set`,  
   store the result in a variable called `common_names`, and print it.  
   *Hint: Use the `intersection()` method.*

7. Find **all unique names** in `name_tuple` and `name_set`,  
   store the result in a variable called `all_names`, and print it.  
   *Hint: Use `union()` method.*

8. Find the **difference** between `name_tuple` and `name_set`,  
   store the result in a variable called `name_difference`, and print it.  
   *Hint: Use the `difference()` method.*

Useful resources to learn about Tuple and Set variables:
- https://www.w3schools.com/python/python_tuples.asp
- https://www.w3schools.com/python/python_sets.asp
---



```python
# Write your code here
name_tuple = ("Alex", "Jane", "Sam", "Sophia")

print(name_tuple[0].upper())

name_set = {"Alex", "Maria", "John", "Maria", "Sophia", "Alex"}
print(name_set)

sorted_names = sorted(name_set)
print(sorted_names)

print(type(name_tuple))
print(type(name_set))
print(type(sorted_names))

common_names = set(name_tuple).intersection(name_set)
print(common_names)

all_names = set(name_tuple).union(name_set)
print(all_names)

name_difference = set(name_tuple).difference(name_set)
print(name_difference)
```

### **Exercise 5**: Arithmetic Operations and Nested `if` Statements

In this exercise, you will practice working with **numbers**, **mathematical operations**, and **conditional statements**, including **nested `if` blocks**.

---

**Instructions**

1. Write a program to ask the user to enter a number and store it in a variable called `user_number`.  
   *Hint: Use the `input()` function.*

2. Convert user_number to an integer, and reassign the result back to `user_number`. (**Don't create a new variable**)

3. Write a program that checks whether `user_number` is **greater than 0**, **less than 0**, or **equal to 0**.

   - If `user_number` is greater than 0, print `"The number is positive"`.
   - If `user_number` is less than 0, print `"The number is negative"`.
   - If `user_number` is equal to 0, print `"The number is zero"`.

4. In the case that `user_number` is **greater than 0**, after printing `"The number is positive"`,  
   add another `if` statement **inside this condition** to check whether the number is **even or odd**:

   - If the number is even, print `"The number is even"`.
   - If the number is odd, print `"The number is odd"`.

   *Hint: Use the modulo operator `%` to check even and odd numbers.*
  **Important:** Pay close attention to indentation when writing nested `if` statements in Python.

Useful resources to learn about arithmetic operations and `if` statements:
- https://www.w3schools.com/python/python_operators_arithmetic.asp
- https://www.w3schools.com/python/python_conditions.asp

---

```python
# Write your code here
user_number = input("Enter a number: ")

user_number = int(user_number)

if user_number > 0:
    print("The number is positive")
    if user_number % 2 == 0:
        print("The number is even")
    else:
        print("The number is odd")

elif user_number < 0:
    print("The number is negative")

else:
    print("The number is zero")
```
