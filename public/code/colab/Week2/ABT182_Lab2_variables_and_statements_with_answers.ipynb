**Exercise 3**

- We have a list of recorded temperatures for Davis. `temp = [65, 69, 58, 82, 53, 84, 64]`
- Calculate/find and print the following:
    - mean of temp. print it with 2 decimal places.
    - min of temp
    - max of temp
    

Useful resources:
- https://www.w3schools.com/python/ref_string_format.asp
- https://www.w3schools.com/python/python_string_formatting.asp
- https://www.geeksforgeeks.org/formatted-string-literals-f-strings-python/


- https://realpython.com/python-min-and-max/

````python
# add your code here
temp = [65, 69, 58, 82, 53, 84, 64]

# take the average
average = sum(temp) / len(temp)

# print the average with 2 decimals
print(f"average temperature is {average:.2f}.")
# .2 sets the precision to 2
# f displays the number as a fixed-point number

# find the min temp and print it
min_temp = min(temp)
print(f"min temp: {min_temp}.")

# find the max temp and print it
max_temp = max(temp)
print(f"max temp: {max_temp}.")
````

**Exercise 4**

- Let's do the exercise above using the methods available in `statistics`  package.
- You need to first import the package `import statistics`, then use proper methods to calculate descriptive statistics for the temp list.
    - mean
    - median
    - mode
    - standard deviation
    - variance


Useful resources:
- https://docs.python.org/3/library/statistics.html
- https://www.w3schools.com/python/module_statistics.asp


````python
import statistics

mean = statistics.mean(temp)
median = statistics.median(temp)
mode = statistics.mode(temp)
st_dev = statistics.stdev(temp)
variance = statistics.variance(temp)

print(f"Mean: {mean: .2f}")
print(f"Median: {median}")
print(f"Mode: {mode}")
print(f"Standard Deviation: {st_dev: .2f}")
print(f"Variance: {variance: .2f}")
````

````python
# Bonus question
my_pass = 'Abt/Hyd 182'
max_attempts = 3
attempts = 0

while attempts < max_attempts:
    password = input("enter the password for this class: ")

    if password == my_pass:
        print("The pass is correct!")
        break
    else:
        attempts += 1

        if attempts == max_attempts:
            print("The account is locked")
            print(f"number of attempts: {attempts}\n")

        else:
            print("The pass is incorrect! Try again.")
            print(f"number of attempts: {attempts}\n")

````

````python
# Import the required module
# remember our analogy: we import a module (CountryInfo) from a package called countryinfo
# ==> bringing only a chapter (CountryInfo) of a book (countryinfo) to your working desk

from countryinfo import CountryInfo
````

````python
# Input the Name of the Country
name = input("Enter your country: ")

# Passing the country name entered by the user to the in-built function of the module CountryInfo
country = CountryInfo(name)
````

````python
# Printing all the details using the Pre-defined Functions

# 1) Printing the Capital of the Country
print("Capital is : ", country.capital())

# 2) Printing the Currency used in the Country
print("Currencies is :", country.currencies())

# 3) Printing the Languages spoken in the Country
print("Language is : ", country.languages())

# 4) Printing the Borders of the Country and Function
# returns bordering countries (ISO3) for a specified country
print("Borders are : ", country.borders())

# 5) Printing the Provinces of the Country
print("Provinces are : ", country.provinces())

# 6) Printing the Areas in the Country
print("The area is : ", country.area())

# 7) Printing the Country Code of the Country
print("Calling code : ", country.calling_codes())

# 8) Printing the Latitudes and Longitudes of the Country
print("Capital Latitudes and Longitudes are : ",
                          country.capital_latlng())

# 9) Printing the TimeZone in the Country
print("TimeZone : ", country.timezones())

# 10) Printing the Current Population in the Country
print("Population : ", country.population())

# 11) Printing the Other Names of the Country
print("Others names : ", country.alt_spellings())

print("Wikipedia page: ", country.wiki())
````

  - Create a dictionary called `city_population` where:
  - keys are **city names**
  - values are their **population**

  Example:
  ```python
  city_population = {
      "Davis": 66799,
      "San Francisco": 815201,
      "Los Angeles": 3898747
  }
```
- Write a program that:
  - Asks the user to enter a city name
  - Checks if the city exists in the dictionary
    - If the city exists:
      - Print the population of the city
      - Print `"Large city"` if the population is greater than 500,000
      - Print `"Small/Medium city"` otherwise
    - If the city does not exist, print `"City not found"`

**Note:** The population values used in this exercise are not exact and are provided for practice purposes only.



````python
# write your code here
city_population = {
    "Davis": 66799,
    "San Francisco": 815201,
    "Los Angeles": 3898747,
    "Sacramento": 528306
}

city = input("Enter a city name: ")

if city in city_population:
    population = city_population[city]
    print(f"Population of {city}: {population}")

    if population > 500000:
        print("Large city")
    else:
        print("Small/Medium city")
else:
    print("City not found")
````
